```python
from matplotlib import pyplot as plt
import pandas as pd

plt.xkcd()
```

```python
data = pd.read_csv('data.csv')
ages = data['Age']
dev_sal = data['All_Devs']
py_sal = data['Python']
js_sal = data['JavaScript']
```

```python
plt.plot(ages, js_sal,color = 'black',label='Java')

plt.plot(ages, dev_sal, color = 'red', 
         linestyle='--', label='All Devs')

plt.plot(ages, py_sal,color = 'green',label='Python')

o_median = 57287

plt.fill_between(ages, py_sal, alpha=0.25,y2=dev_sal,color='blue', 
                 where=(py_sal>dev_sal),interpolate=True, label='Above Avg')

plt.fill_between(ages, py_sal, alpha=0.25,y2=dev_sal,color='red', 
                 where=(py_sal<=dev_sal),interpolate=True, label='Below Avg')

plt.legend()

plt.title('Median Salary (Rs) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary (Rs)')
#plt.savefig('line53.png')
plt.tight_layout()

plt.show()
```
