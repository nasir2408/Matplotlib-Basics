```python
from matplotlib import pyplot as plt

print(plt.style.available)
plt.style.use('ggplot')
#plt.xkcd()
```

```python
slices = [60, 40, 30, 20]
labels = ['Sixty', 'Forty', 'Thirty', 'Twenty']
colors = ['blue','red', 'yellow', 'green']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})

plt.title('My Awesom Pie Chart')
plt.tight_layout()
#plt.savefig('pie31.png') 
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0.1,0.2,0.3,0.4]

plt.pie(slices, labels=labels, explode=explode, shadow = True, startangle=90, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})

plt.title('My Awesom Pie Chart')
plt.tight_layout()
#plt.savefig('pie35.png')
plt.show()
```
