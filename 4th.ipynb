```python
from matplotlib import pyplot as plt

#plt.style.use('fivethirtyeight')
plt.xkcd()
```

```python
m = [1, 2, 3, 4, 5, 6, 7, 8, 9]

p1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
p2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
p3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

l= ['d1', 'd2', 'd3']
col = ['green', 'red', 'blue']
#plt.pie([1, 1, 1], labels=['p1', 'p2', 'p3'])
plt.stackplot(m, p1, p2, p3, labels=l, colors = col, )

#plt.legend(loc ='lower left')
plt.legend(loc =(0.07, 0.05))

plt.title('My Awesom Stack Plot')
plt.tight_layout()
#plt.grid()
plt.savefig('stack42.png')
plt.show()
```
