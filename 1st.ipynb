```python
#Simple line plot

data = pd.read_csv('data1.csv')
x = data['Age']
ds = data['All_Devs']
ps = data['Python']
js = data['JavaScript']

fig1, ax1= plt.subplots()
fig2, ax2= plt.subplots()


ax2.plot(x, ds,  color='#5a7d9a', linewidth=3, label= 'All_a')
ax2.plot(x, ps,  color='#adab3b', linewidth=3, label= 'All_b')
ax1.plot(x, js, color='#444444', linestyle='--', label= 'All_y')

#ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (Rs)')
ax1.set_title('Median Salary (Rs) by Ages')
ax1.legend()
#plt.legend(['All_y', 'All_a'])

ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (Rs)')
#ax2.set_title('Median Salary (Rs) by Ages')
ax2.legend()

#plt.grid(True)

plt.tight_layout()

#plt.savefig('plot11.png') #save the plot 

plt.show()

#fig1.savefig('fig1.png')
#fig2.savefig('fig2.png')
```
